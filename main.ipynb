```python
import re
from matplotlib import pyplot as plt
from wordcloud import WordCloud
from os import listdir

%matplotlib inline 
```

```python
file = open('diarios','r')
text = file.read()
file.close()
```

```python
#split a document to words
text_1 = re.split(r'[\n.()!\\\';:\-, ""/]',text.lower())
```

```python
# delete some empty string
text_2 = [x for x in text_1 if len(x) > 0]
```

```python
# get word frequency and delete some stopwords
wg = text_2
wd = {}
```

```python
for w in wg:
    if w in stopwords:
        continue
    else:
        str(w)
        if w not in wd:
            wd[w] = 1
        else:
            wd[w] += 1
```

```python
# get the wordcloud
wordcloud = WordCloud(width=1600, height=800).generate_from_frequencies(wd)
```

```python
plt.figure(figsize=(30,50))
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```
